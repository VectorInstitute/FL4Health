```python
dict_averages = {}
dict_std = {}
for categories_key,categories_value in dict_categories.items():
    cv = categories_value.split('\n')
    for i in range(0, len(cv)-1, 2):
        cv_avg = cv[i].split(':')
        cv_std = cv[i+1].split(':')
        if i < 12:
            keeey = cv_avg[0][0:8]
        else:
            keeey = cv_avg[0][-18:]
        if keeey in dict_averages.keys():
            pass
        else:
            dict_averages[keeey] = []
            dict_std[keeey] = []
        dict_averages[keeey].append(float(cv_avg[1][1:]))
        dict_std[keeey].append(float(cv_std[1][1:]))
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))
```

```python
# Create an array of x values for each category
x = np.arange(len(dict_categories.keys()))
print(dict_averages.keys())
for category in dict_averages.keys():
    
    sorted_categories, sorted_avg, sorted_std = zip(*sorted(zip(dict_categories.keys(), dict_averages[category],dict_std[category]), key=lambda x: x[1], reverse=False))
    # Create the bar chart for averages
    plt.bar(x, sorted_avg, yerr=sorted_std, capsize=5, align='center', alpha=0.7, label='Average')

    # Set the x-axis labels
    plt.xticks(x, sorted_categories, rotation=90)

    # Set labels and title
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.title(f'Bar Chart with Average and Standard Deviation for : {category}')

    # Show the legend
    plt.legend()

    # Show the plot
    plt.show()
```
